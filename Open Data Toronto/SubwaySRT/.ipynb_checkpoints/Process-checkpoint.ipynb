```python
# import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import os

pathway1 = os.path.dirname(os.path.abspath('Process.ipynb'))
csv_to_load = os.path.join(pathway1, "subway_master.csv")

df_master = pd.read_csv(csv_to_load)
df_master.describe()
```

```python
print(df_master.shape[0], df_master.shape[1])
```

```python
df_master.columns
```

```python
df_master.dtypes
```

```python
df_master.head(5)
```

```python
df_date = df_master.groupby('Date').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_date.head()
```

```python
df_year_month = df_master.groupby(["Year", "Month"]).agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_year_month.head()
```

```python
df_year = df_master.groupby("Year").agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_year.head()
```

```python
df_day = df_master.groupby("Day").agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_day = df_day.reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
df_day
```

```python
df_code = df_master.groupby('Code').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_code.head()
```

```python
count = df_code["Min Delay"]["count"]
mean1 = df_code["Min Delay"]["mean"]
sum1 = df_code["Min Delay"]["sum"]
mean2 = df_code["Min Gap"]["mean"]
sum2 = df_code["Min Gap"]["sum"]

dict1 = {"Count": count, 
        "Min Delay - Mean": mean1, 
        "Min Delay - Sum": sum1, 
        "Min Gap - Mean": mean2, 
        "Min Gap - Sum": sum2}

df_sort_code = pd.DataFrame(dict1)

df_sort_code = df_sort_code.sort_values("Count", ascending=False)
df_sort_code.head()
```

```python
df_station = df_master.groupby('Station').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_station.head()
```

```python
df_vehicle = df_master.groupby('Vehicle').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_vehicle.head()
```

```python
df_mean_temp = df_master.groupby('Mean Temp').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_mean_temp.head()
```

```python
df_precip = df_master.groupby('Precipitation').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_precip.head()
```

```python
df_rain = df_master.groupby('Total Rain (mm)').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_rain.head()
```

```python
df_snow = df_master.groupby('Total Snow (cm)').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_snow.head()
```

```python
df_snow_ground = df_master.groupby('Snow on Grnd (cm)').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_snow_ground.head()
```

```python
df_wind = df_master.groupby('Spd of Max Gust (km/h)').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_wind.head()
```

```python
df_holiday = df_master.groupby('holiday').agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_holiday
```

```python
count = df_holiday["Min Delay"]["count"]
mean1 = df_holiday["Min Delay"]["mean"]
sum1 = df_holiday["Min Delay"]["sum"]
mean2 = df_holiday["Min Gap"]["mean"]
sum2 = df_holiday["Min Gap"]["sum"]

dict1 = {"Count": count, 
        "Min Delay - Mean": mean1, 
        "Min Delay - Sum": sum1, 
        "Min Gap - Mean": mean2, 
        "Min Gap - Sum": sum2}

df_holiday = pd.DataFrame(dict1)

df_holiday = df_holiday.sort_values("Count", ascending=False)
df_holiday
```

```python
df_bound_line = df_master.groupby(['Bound', 'Line']).agg({'Min Delay': ['count', 'mean', 'sum'], 'Min Gap': ['mean', 'sum']})
df_bound_line
```

```python
# Code Groupby
fig, ax = plt.subplots()

# Example data
codes = df_sort_code["Count"].index.tolist()
y_pos = np.arange(len(codes))
count = df_sort_code["Count"].tolist()

ax.barh(y_pos, count, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(codes)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Number of Codes from 2014-2019')
ax.set_title('TTC Subway Codes')

plt.show()
```
